# EDA OCEAN (Big 5)

### 1. Concaténation des datasets Essays et Ocean Synthetic

```python
import pandas as pd
import numpy as np

import plotly.express as px
import plotly.graph_objects as go
import matplotlib.pyplot as plt
```

```python
df_essays_train = pd.read_csv("essays_train.csv")
df_essays_test = pd.read_csv("essays_test.csv")
df_essays_validation = pd.read_csv("essays_validation.csv")

df_essays = pd.concat([df_essays_train, df_essays_test, df_essays_validation], ignore_index=True)
```

```python
df_essays.info()
```

```python
df_essays_renamed = df_essays.rename(columns={
    'O': 'Openness',
    'C': 'Conscientiousness',
    'E': 'Extraversion',
    'A': 'Agreeableness',
    'N': 'Neuroticism',
    'text': 'Text'
})

df_essays_cleaned = df_essays_renamed.drop(columns=['ptype', '__index_level_0__'])
```

```python
df_ocean = pd.read_csv("OCEAN-synthetic.csv")

df_ocean.info()
```

```python
df_ocean.head()
```

```python
df_ocean['Agreeableness'].fillna(df_ocean['Agreeableness'].mean(), inplace=True)
df_ocean['Neuroticism'].fillna(df_ocean['Neuroticism'].mean(), inplace=True)
```

```python
df_ocean.info()
```

```python
df_ocean_binarized_median = df_ocean.copy()

ocean_cols = ['Openness', 'Conscientiousness', 'Extraversion', 'Agreeableness', 'Neuroticism']

for col in ocean_cols:
    median_val = df_ocean_binarized_median[col].median()
    df_ocean_binarized_median[col] = (df_ocean_binarized_median[col] >= median_val).astype(int)
```

```python
df = pd.concat([df_essays_cleaned, df_ocean_binarized_median], ignore_index=True)
```

```python
df.to_csv("concat_ocean.csv", index=False)
```

```python
df.info()
```

```python
traits = ["Openness", "Conscientiousness", "Extraversion", "Agreeableness", "Neuroticism"]

# Création des colonnes 'Profile' et 'Length_text'

df['Profile'] = df[traits].astype(str).agg(''.join, axis=1)
df["Length_text"] = df["Text"].apply(len)

df.head()
```

### 2. Distribution des 32 profils OCEAN (5 bits)

```python
profile_counts = df['Profile'].value_counts()

fig = px.bar(
    profile_counts,
    title='Distribution des 32 profils OCEAN (5 bits)',
    labels={'index': 'Profils', 'value': 'Count'},
    color=profile_counts.index,
    color_discrete_sequence=px.colors.qualitative.Pastel
)

fig.show()
```

### 3. Distribution de chaque trait OCEAN

```python
for col in traits:
    trait_counts = df[col].value_counts().reset_index()
    trait_counts.columns = ['Value', 'Count']

    fig = px.pie(
        trait_counts,
        names='Value',
        values='Count',
        title=f'Distribution du trait: {col}',
        color_discrete_sequence=px.colors.qualitative.Pastel,
        labels={'Value': f'{col} (0=Faible, 1=Élevé)', 'Count': 'Nombre'}
    )

    fig.show()
```

### 4. Matrice de corrélation entre les traits OCEAN

```python
corr_matrix = df[traits].corr()

fig = px.imshow(
    corr_matrix,
    text_auto=True,
    color_continuous_scale=px.colors.sequential.Teal,
    title='Matrice de corrélation des traits OCEAN',
    labels=dict(x="Trait 1", y="Trait 2", color="Corrélation"),
    width=800,
    height=800
)

fig.update_xaxes(side="top")

fig.show()
```

### 4. Distribution des longueurs de texte

```python
fig = px.histogram(
        df,
        x='Length_text',
        nbins=50,
        title='Distribution des longueurs de texte',
        labels={'Length_text': 'Longueur du texte'},
        marginal="box",
    	color_discrete_sequence=px.colors.qualitative.Pastel)

fig.show(color_discrete_sequence=px.colors.qualitative.Pastel)
```

#### Longueur du texte par trait BIG5

```python
color_discrete_map = {
    0: px.colors.qualitative.Pastel[0],
    1: px.colors.qualitative.Pastel[1]
}

for col in traits:
    fig = px.box(
        df,
        x=col,
        y='Length_text',
        title=f'Longueur du texte par niveau de {col}',
        labels={col: f'{col} (0=Faible, 1=Élevé)', 'Length_text': 'Longueur du texte'}, 
        color=col,
        color_discrete_map=color_discrete_map,
        points="all"
    )
    fig.show()
```
